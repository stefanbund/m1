### A four day sample: mean surge as predictor to mp surge
Given a four day sample, the mp to mean relationship holds, only in very significant increases in mp value. It appears that high volumes of buy orders, placed during low points of mp value, account for the increase in mean. But the same precursor holds for the mean-to-mp relationship. When you observe maximum increase in mean value (surges), there is an orderly increase in mp, but with some delay. In the case of the four day, a ~10% increase in mean foreshadowed the same increase in mp value. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
#1-6-post-hft-AVAX/jul29 24 hr/Thu Jul 28 2022 06_18_25 GMT-0700 (Pacific Daylight Time)-MEANSHIFT.csv
csv = pd.read_csv("Thu Jul 28 2022 06_18_25 GMT-0700 (Pacific Daylight Time)-MEANSHIFT (1).csv")
#csv.head
```

### Inversely Correlated Mean and Median Values, in the Subsample
Subsampling mean and mp values, yields a negative correlation. 

```python
X = csv[['mean']]
Y = csv['mp']
regr = linear_model.LinearRegression()
regr.fit(X,Y)
print(regr.coef_)
```

however, there is agreement that higher means are associated with higher median prices. However, an event based predictor, or precursor association would need to be established to justify monitoring mean/median. 

This is not a time series, only an association between mean order book value and mp, or price outcome. This is a relationship that seems to generalize, 'weak mean associates to strong mp.' **Chi square association opportunity.** 

**In general, higher mean values associate with higher mp.**

```python
res = sns.lineplot(x="mean", y="mp", data=csv)
plt.show()
```

### Exploration 1: Time Series Analysis: Seeking Evidence of Precursor Events to Price
As a potential precursor to price rising, a dramatic surge in the mean value of orders appears.
mp shows some response on the timeseries, after a surge in mean value. 

**There is some evidence that the accumulation of mean value acts as a precursor to increased mp.** 

```python
res = sns.lineplot(x="timeStamp", y="mp", data=csv)
#import datetime
#ml = str(datetime.datetime.now) + "-mp-line.png"
plt.savefig("mp-line.png")
plt.show()
```

Here is a surge in mean values, pushing up the value of orders. On the same time series as median (settled price) above, the surge in mean values acts as an event precursor to the surge in median price. There is also a collapse in mean, which is reflected in the deflated value of mp.

**Clearly, the time series creates advantages in visualizing the surge in mean as a time-based precursor to the rise in the median.**

Mean Time Series Plot

```python
res = sns.lineplot(x="timeStamp", y="mean", data=csv)
plt.savefig("mean-line.png")
plt.show()
```
